## AIC - BIC

```python
liste_patients = [14, 15, 16, 17, 18, 20, 21, 22, 23, 25, 26, 28, 29, 32, 33, 34, 36, 37, 38]

#on enleve le patient 20 car il n as pas assez de données
liste_patients.remove(20)
```

```python
def selection_nombre_clusters(data, patient):
    
    #Normalisation des données
    scaler = StandardScaler()
    data_scaled = scaler.fit_transform(data)

    #Réduction de dimenison - PCA
    pca = PCA(2)   #2 dimensions pour la visualisation
    data_pca = pca.fit_transform(data_scaled)

    aic = []
    bic = []
    K = range(1, 15) # Nombre de clusters à tester
    # Sélection du nombre de clusters avec AIC et BIC
    for k in K:
        gmm = GaussianMixture(n_components=k, random_state=0)
        gmm.fit(data_pca)  # X_pca = gènes projetés sur 2-10 dimensions
        aic.append(gmm.aic(data_pca))
        bic.append(gmm.bic(data_pca))

    plt.plot(K, aic, label='AIC', marker='o')
    plt.plot(K, bic, label='BIC', marker='s')
    plt.xlabel('Nombre de clusters')
    plt.ylabel('Score')
    plt.title('Sélection du nombre de clusters (GMM) patient {}'.format(patient))
    plt.legend()
    plt.grid()
    plt.show()
```

```python
for num_patient in liste_patients:
    patient = patients(num_patient, bt)
    patient = patient.T.astype('float')
    print(f"Patient {num_patient}")
    selection_nombre_clusters(patient, num_patient)
```
